### Two Sum

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        i = 0
        d = {}

        for i in range(len(nums)):
            bal = target - nums[i]
            if bal in d:
                return [d[bal], i]
            d[nums[i]] = i
```

### Longest Common Prefix

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        i = 0
        com_prefix = ""

        while True:
            if i < len(strs[0]):
                curr = strs[0][i]
            else:
                return com_prefix

            for s in strs[1:]:
                if not (len(s) > i and curr == s[i]):
                    return com_prefix

            com_prefix += curr
            i += 1
```

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        if not strs:
            return ""
        
        # zip(*strs) pairs characters by index: (char1, char2, char3...)
        for i, chars in enumerate(zip(*strs)):
            # If the set of characters at this index > 1, they aren't all the same
            if len(set(chars)) > 1:
                return strs[0][:i]
        
        # If the loop finishes, the shortest string is the prefix
        return min(strs, key=len)
```

```python
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        k = 0
        for i in range(len(nums)):
            if nums[i] != val:
                nums[k] = nums[i]
                k += 1
        return k
```

### 35. Search Insert Position

```python
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        for i in range(len(nums)):
            if nums[i] == target \
            or nums[i] > target:
                return i

        return len(nums) 
```

### 66. Plus One

```python
class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:         
        digits_str = [str(i) for i in digits]
        ans = str(int("".join(digits_str)) + 1)
        ans = [int(i) for i in ans]
        return ans

# the above is not recommended for large numbers
# the below is recommended from gpt
class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        n = len(digits)
        
        # Start from the last digit
        for i in range(n - 1, -1, -1):
            if digits[i] < 9:
                digits[i] += 1
                return digits
            digits[i] = 0  # carry over
        
        # If all digits were 9
        return [1] + [0] * n
```

### 88. Merge Sorted Array

```python
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        merge = []
        l, r = 0, 0
        while l < m and r < n:
            if nums1[l] <= nums2[r]:
                merge.append(nums1[l])
                l += 1
            elif nums1[l] > nums2[r]:
                merge.append(nums2[r])
                r += 1
        while l < m:
            merge.append(nums1[l])
            l += 1
        
        while r < n:
            merge.append(nums2[r])
            r += 1

        for i in range(len(merge)):
            nums1[i] = merge[i]

        return merge
```
